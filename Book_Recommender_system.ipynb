Building popularity based recommendation system

Here we will take top 80 books which have their avg rating good which have atleast 200 no of ratings 



```python
#merge ratings and books on top of isbn because it is in both files
ratings_with_name = ratings.merge(books,on='ISBN')
```

```python
ratings_with_name.shape
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
avg_rating_df.describe()
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on='Book-Title')
popular_df
```

```python
popular_df['num_ratings'].describe()
```

```python
popular_df = popular_df[popular_df['num_ratings']>=200].sort_values('avg_ratings',ascending = False).head(85)
```

```python
popular_df = popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings','Year-Of-Publication']]
```

```python
popular_df
```

Building Collaborative Filtering Based Recommendation System


```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
users_who_rate = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(users_who_rate)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
 #final-ratings contains the book which have been rated by 
 #more than 50 users and have been rated by thoseusers who have rated atleast 200 books
```

```python
pt = final_ratings.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores[0]   #showing the score of 1st book with other books, in terms of eucledian distance
```

```python
similarity_scores.shape
```

```python
# defining a function which will show 6 books on giving one name
def recommend(book_name):
  #index fetch
  index = np.where(pt.index==book_name)[0][0]
  similar_items = sorted(list(enumerate(similarity_scores[index])),key = lambda x:x[1], reverse = True)[1:7]
  
  data = []
  for i in similar_items:
    item = []
    temp_df = books[books['Book-Title']==pt.index[i[0]]]
    item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
    item.extend((temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
    item.extend((temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
    
    data.append(item)

  return data
```

```python
recommend('A Map of the World')
```
